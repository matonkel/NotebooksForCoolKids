## What is required?

1.) ArcGIS Enterprise with Admin or Publisher Rights
2.) ArcGIS Pro

## What is included in ArcGIS Enterprise

* Portal for ArcGIS
* ArcGIS Server
    * Federated
    * Stand-alone
* ArcGIS Data Store
* ArcGIS Web Adaptor

## Web tools & Geoprocessing services

* Web Tool
    * Publish a set of geoprocessing tools from ArcGIS Pro to and ArcGIS Enterprise federated server
* Geoprocessing service
    * Publish a set of geoprocessing tools to stand-alone ArcGIS Server from ArcMap or ArcGIS Pro 2.4 and above
* Web Tool
    * = Portal Item + Geoprocessing service

## Workflow

### 1.) Design & Author

* Tools are first authored and run in ArcGIS Pro
    * System tools
    * Custom tools created using ModelBUilder or Python
* Authoring considerations
    * Documentation is a must to have tool understood and used
    * Performance
        * Write tool intermediate data to 'in_memory' workspace
        * Use layers for project data instead of disk path
        * Preprocess data if possible
    
### 2.) Publish

* Account Role
    * Need Admin account to Server or
    * Publisher need to be configured to share web tools
* Execution
    * Synchronous or asynchronous
        * Synchronous is quicker
* Input
    * Input areprovied by tool
        * How?
* Data
    * Is the local data copied over or refrenced?
    * Server needs to acess the data used by the tool
* Output
    * IS the output saved on client or Server side?
    * Option to have server or client draw results from tool

* Web Tool Publish
    * From Pro

* Config settings for web tools
    * Execution mode
        * Synchronous
            * General rule for web tools with small output
        * Asynchronous
            * Map Image
            * Feature Service Ouput
    * Parameter Types
        * User-defined
            * Client provided

        * Constant

    * Output

* Geoprocessing Services Publish
    * Account Role
        * Must have Server Admin
    * Must have ArcGIS Server Connection
        * With and administrator connection type
        * Right click connection and publish as geoprocessing service
    
    



### 3.) Consume

### Benifits to web share

* Increase reach of spatial analysis
    * Adds powerful analysis functionality to web apps
    * Shares analysis from GIS pros with other
    * Can be used everywhere by any device

* Simplifiy GIS Anlysis
    * Web app Audience does not need GIS Expertise
    * GIS software does not need to be installed locally
    * Programming is not required to creat or use web tools
